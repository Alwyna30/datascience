```python
# Checking for missing values
print(df.isnull().sum())

# Handling missing values - Imputation with mean or median
df.fillna(df.mean(), inplace=True)
# Alternatively, you could remove rows with missing values
df.dropna(inplace=True)
```

```python
print(df.dtypes)
```

```python
# IQR method
Q1 = df.quantile(0.25) 
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Filtering out the outliers
df_no_outliers = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Summary of statistics
stats_summary = df.describe()

# Median for each column
median_values = df.median()

# Interquartile Range (IQR)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

print(stats_summary)
print("Median values:\n", median_values)
print("IQR:\n", IQR)
```

Brief Report on Cardiographic Dataset Findings
1. Objective
The goal of the analysis was to explore fetal heart rate (FHR) data to uncover patterns, identify correlations, and gain insights into fetal well-being.

2. Key Findings
Baseline FHR (LB): The average fetal heart rate was around 132 bpm, with most values showing a normal distribution.
Variability in Heart Rate: High variability in short-term (ASTV, MSTV) and long-term (ALTV, MLTV) measures was observed. These measures highlight cases of abnormal fetal heart patterns.
Accelerations and Decelerations: The dataset exhibited a moderate number of accelerations (AC) and decelerations (DL, DS, DP), suggesting variability in fetal well-being.
Outliers: Significant outliers were detected in ALTV and MLTV, potentially indicating distressed fetal conditions in some cases.
Correlations: A negative correlation between baseline FHR and long-term variability (ALTV) suggests that higher variability may correspond to lower baseline heart rate, which could indicate potential risk.
3. Conclusion
The dataset revealed critical patterns in fetal health, particularly concerning variability in heart rate and the presence of decelerations. Further analysis could focus on understanding the relationship between uterine contractions, heart rate variability, and fetal outcomes.
